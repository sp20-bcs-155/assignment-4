```python

from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score,recall_score,precision_score
from sklearn.model_selection import train_test_split
from sklearn import metrics, model_selection
from sklearn import preprocessing
import pandas as pd
import numpy as np
from sklearn.model_selection import LeavePOut
from  sklearn.model_selection import ShuffleSplit 
from sklearn.model_selection import cross_val_score
```

```python
df = pd.read_csv('gender-prediction.csv')
```

```python
x = df.drop('gender', axis=1)
y = df['gender']
```

```python
labels = preprocessing.LabelEncoder()
beard_encoded = labels.fit_transform(x.beard)
hair_encoded = labels.fit_transform(x.hair_length)
scarf_encoded = labels.fit_transform(x.scarf)
eye_encoded = labels.fit_transform(x.eye_color)
gender_encoded = labels.fit_transform(y)
```

```python
features = list(zip(x.height,x.weight,beard_encoded,hair_encoded,x.shoe_size,scarf_encoded,eye_encoded))
target = gender_encoded
```

```python
X_train, x_test, Y_train, y_test = train_test_split(features, target, test_size = 0.33, random_state = 5)
```

```python
model = RandomForestClassifier()
model.fit(X_train,Y_train)
prediction = model.predict(x_test)
model_acc = accuracy_score(y_test, prediction)*100
print("ACCURACY = ",model_acc)
model_cm = metrics.confusion_matrix(y_test, prediction)
print(model_cm)
```

```python
model = SVC()
model.fit(X_train,Y_train)
prediction = model.predict(x_test)
model_acc = accuracy_score(y_test, prediction)*100
print("ACCURACY = ",model_acc)
model_cm = metrics.confusion_matrix(y_test, prediction)
print(model_cm)
```

```python
model = MLPClassifier()
model.fit(X_train,Y_train)
prediction = model.predict(x_test)
model_acc = accuracy_score(y_test, prediction)*100
print("ACCURACY = ",model_acc)
model_cm = metrics.confusion_matrix(y_test, prediction)
print(model_cm)
```
